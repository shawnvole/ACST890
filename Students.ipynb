```python
# inmport packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv('~/Documents/student_results.csv') #read csv
dataset.head() #display the first 5 lines of dataset
```

```python
dataset = dataset.replace("NA",0) #remove the NAs from the table
dataset = dataset.replace(np.nan,0) #remove the NaN(missing values) from the table
dataset.head()
```

```python
weights = np.array([[0.07],
                   [0.03],
                   [0.03],
                   [0.02],
                   [0.03],
                   [0.02],
                   [0.024],
                   [0.016],
                   [0.07],
                   [0.03],
                   [0.04],
                   [0.02],
                   [0.6]])
print(weights.sum()) #Check for the weights adding up to 1 or not
dataset['Total.marks'] = np.dot(dataset.iloc[:,1:14], weights) #Matrix Multiplication
dataset.head()
```

```python
import seaborn as sns
ax = sns.boxplot(x="Student category", y="Total.marks", data=dataset) #plot the boxplot by student category
```

```python
# create a table for mean, variance and SD of Total Marks by Class No.
stat_table = dataset.groupby('Class No').agg({"Total.marks": ['mean', 'var', 'std']})
stat_table
```

```python
# assign the grades to each student
Grade = []

for row in dataset['Total.marks']:
    if row < 50:
        Grade.append('F')
    elif row < 65:
        Grade.append('P')
    elif row < 75:
        Grade.append('Cr')
    elif row < 85:
        Grade.append('D')
    else:
        Grade.append('HD')
        
dataset['Grade'] = Grade
dataset.head()
```

```python
# Linear regression for one x variable
import statsmodels.api as sm 

X = dataset["Assessment 1"] 
y = dataset["Final exam"]
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

```python
# fit plot for linear regress
fig, ax = plt.subplots()
ax.plot(dataset["Assessment 1"], predictions, color='red')
ax.scatter(dataset["Assessment 1"], dataset["Final exam"])

fig.show()
```

```python
# Linear regression for multiple x variables
X = dataset.iloc[:,[1,9,3,8]]
y = dataset["Final exam"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

```python
# assign status to each student
Status = []

for row in dataset['Grade']:
    if row == 'F':
        Status.append('incomplete')
    else:
        Status.append('complete')
        
dataset['Status'] = Status
dataset.head()
```

```python
# logistic regression for training data with first 100 (0:99) students
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
y_train = pd.get_dummies(dataset.loc[0:99,'Status'])['complete']
X_train = dataset.loc[0:99,'Assessment 3']
X_train = sm.add_constant(X_train)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# test the model with the remaining data points
y_test = pd.get_dummies(dataset.loc[100:,'Status'])['complete']
X_test = dataset.loc[100:,'Assessment 3']
X_test = sm.add_constant(X_test)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```
